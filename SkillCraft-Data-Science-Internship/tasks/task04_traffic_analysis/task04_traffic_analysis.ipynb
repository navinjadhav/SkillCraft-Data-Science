**Goal:** Analyze patterns in daily accident counts and compare by weather/day.

Dataset: `data/traffic_accidents_2025_synthetic.csv` (synthetic, offline)


```
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = "../../data/traffic_accidents_2025_synthetic.csv"
df = pd.read_csv(DATA_PATH)
df["date"] = pd.to_datetime(df["date"])
df.head()
```

## 1) Trend over time

```
daily = df.sort_values("date")

plt.figure(figsize=(10,4))
plt.plot(daily["date"], daily["accidents"])
plt.title("Daily Accidents (2025)")
plt.xlabel("Date")
plt.ylabel("Accidents")
plt.tight_layout()
plt.show()
```

## 2) Average accidents by day of week

```
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
by_day = df.groupby("day")["accidents"].mean().reindex(order)
by_day
```

```
plt.figure(figsize=(8,4))
plt.bar(by_day.index, by_day.values)
plt.title("Average Accidents by Day of Week")
plt.xlabel("Day")
plt.ylabel("Avg accidents")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

## 3) Average accidents by weather

```
by_weather = df.groupby("weather")["accidents"].mean().sort_values(ascending=False)
by_weather
```

```
plt.figure(figsize=(6,4))
plt.bar(by_weather.index, by_weather.values)
plt.title("Average Accidents by Weather")
plt.xlabel("Weather")
plt.ylabel("Avg accidents")
plt.tight_layout()
plt.show()
```

## Summary
- Accident counts vary by day of week and weather.
- Bad weather (rain/storm) shows higher average accidents in this dataset.
